# Specific columns dataframe

```python
data1=pd.read_csv('CRM.csv',usecols=[0])
data1
```

# Skipping top row or Setting top row as the title 

```python
data1=pd.read_csv('CRM.csv',skiprows=1)
data1
```

# Specific row skipping

```python
data1=pd.read_csv('CRM.csv',skiprows=[2])
data1
```

```python
data1=pd.read_csv('CRM.csv',skiprows=[2,6,7])
data1
```

# index_col,  Making a specific column as index

```python
data1=pd.read_csv('CRM.csv',index_col='speed')
data1
```

# header, Making a specific row a as a header.

```python
data1=pd.read_csv('CRM.csv', header=1)
data1 
```

# None header

```python
data1=pd.read_csv('CRM.csv', header=None)
data1 
```

# Set header by own name using prefix

```python
data1=pd.read_csv('CRM.csv',header=None, prefix='Column')
data1 
```

# Giving specific Name of a column

```python
data1=pd.read_csv('CRM.csv', names=['a','b'])
data1 
```

```python
data1=pd.read_csv('CRM.csv')
data1.head(10)
```

```python
data1.tail(10)
```

# Changing datatype of a column

```python
data1=pd.read_csv('CRM.csv',dtype={'speed':'float','risk':'float'})
data1 
```

# replacing values with true or false 

```python
data1=pd.read_csv('CRM.csv',true_values=['yes'],false_values=['no'])
data1
```
